### Description: 
> The TikTok video from the account @smartypants shows a short discussion centered around the well known question of whether water is wet. The video features two presenters standing behind a counter with a clear container of water placed in front of them. One presenter introduces the topic by raising the question directly, asking whether water itself can be classified as wet or if wetness is a property that only applies when a liquid comes into contact with another surface. The other presenter responds by examining the meaning of the word “wet” and how it is used in everyday language. He explains that people usually describe something as wet when a sufficient amount of water clings to or covers its surface. Based on this definition, he argues that water cannot be wet because wetness is a condition created by water acting on something else. The male presenter then challenges this view by suggesting that water molecules are constantly touching one another and therefore are always surrounded by other water molecules. He argues that if being coated in water is what makes something wet, then water itself should qualify because each particle of water is covered by more water. Their debate continues as they alternate between these two perspectives. They bring attention to the difference between a descriptive property that applies to objects, such as a towel or a table becoming wet, and the intrinsic physical identity of water as a liquid. The video frames these two explanations as competing interpretations that depend on how wetness is defined, which encourages viewers to form their own stance and contribute comments that support one side or the other. In doing so, the video creates a straightforward but conceptually interesting question that invites argument based on definitions, observable physical properties, and individual interpretations of language.
    


## **Research Question (this should be clearly scoped):** 
---

Are TikTok commenters more likely to say that water is wet, or that water is not wet?

## **Concepts and Reflection**
--- 

Trial #1: 
    
* Concept 1: Commenter is supportive of the premise that water is wet
* Concept 2: Commenter is generally in disagreement with the premise that water is wet

Krippendorff’s Alpha Scores:
* Concept 1: 0.482
* Concept 2: 0.605

**REVISED CONCEPTS:**
* Concept 1: Comment explicitly states or clearly supports the idea that water is wet
    * A comment receives a 1 for Concept 1 if it clearly expresses agreement with the statement that water is wet. This includes direct assertions, explanations that water can be wet because water molecules adhere to each other, and humorous or sarcastic remarks that still communicate the stance that water is wet. A comment receives a 0 if it does not take a clear stance, if it argues that water is not wet, or if it is unrelated, ambiguous, a tag, a question, or contains only emojis.
* Concept 2: Comment explicitly states or clearly supports the idea that water is not wet
    * A comment receives a 1 for Concept 2 if it clearly states that water is not wet or gives explanations that wetness is a property applied only to external surfaces and not to water itself. This includes direct statements, scientific or pseudo-scientific explanations, and humorous or sarcastic remarks that clearly articulate that water is not wet. A comment receives a 0 if it does not take a clear stance, if it supports water being wet, or if it is unrelated, ambiguous, a tag, a question, or contains only emojis.

* If a comment does not clearly pick a side in the debate, then Concept 1 = 0 and Concept 2 = 0. We won’t force Neutral, unclear, or off-topic comments into one category.

Krippendorff’s Alpha Scores:
* Concept 1: 0.720
* Concept 2: 0.791

Reflection: 
We revised our concepts to increase clarity, mutual exclusivity, and coder agreement. Our original definitions (“supportive of the premise that water is wet” and “generally in disagreement with the premise that water is wet”) were too vague and required subjective interpretation. They did not specify what counts as evidence of support, how to label jokes or scientific explanations, or how to treat neutral or off-topic comments. The lecture emphasized that concepts must be defined so clearly that a young child could reliably code them, and our low original alpha scores indicated that our definitions were not operationalized enough to meet that standard.

The revised definitions directly address this issue by adding explicit coding rules, making each concept dependent on clear linguistic markers, and establishing a consistent rule for handling ambiguous comments. These changes reduce guesswork, eliminate overlap between the two concepts, and ensure all coders treat edge cases in the same way. Now, both human and LM reliability increased.

## **Final LLM Prompt and Inter-Rater Reliability Scores**
---

### **Final LLM Prompt:**
> You are a research assistant helping us analyze TikTok comments for a research project. Your task is to code comments from a TikTok video debating the question “Is water wet?” You will assign two binary variables, Concept 1 and Concept 2, based on the definitions below.

> RESEARCH QUESTION:
Are TikTok commenters more likely to say that water is wet, or that water is not wet?

> CONCEPT 1: Comment explicitly states or clearly supports the idea that water IS wet.
Code Concept 1 as 1 if the comment:
> - Directly says water is wet.
> - States or implies that water can be wet because water molecules cover each other.
> - Gives an explanation that supports water being wet.
> - Uses humor or sarcasm but still clearly communicates the stance that water is wet.
Examples of Concept 1 = 1:
> - “Water is definitely wet.”
> - “Obviously it’s wet. Molecules coat each other.”
> - “Yes water is wet stop fighting.”

> Code Concept 1 as 0 if:
> - The comment does NOT clearly state that water is wet.
> - The comment says water is NOT wet.
> - The comment is neutral, unclear, unrelated, a joke with no stance, a tag, a question, or emoji-only.

> CONCEPT 2: Comment explicitly states or clearly supports the idea that water is NOT wet.
Code Concept 2 as 1 if the comment:
> - Directly says water is not wet.
> - States that wetness is a property that applies only to surfaces, not to water itself.
> - Explains that water makes other things wet but is not wet on its own.
> - Uses humor or sarcasm but still clearly communicates that water is not wet.
Examples of Concept 2 = 1:
> - “Water isn’t wet. It makes things wet.”
> - “Wetness is a property of solids, not liquids.”
> - “Nope water is not wet.”

> Code Concept 2 as 0 if:
> - The comment does NOT clearly state that water is not wet.
> - The comment says water IS wet.
> - The comment is neutral, unclear, unrelated, a tag, a joke with no stance, a question, or emoji-only.

> IMPORTANT:
> A comment may be coded as (0,0) if it does not take a stance.
> A comment should NEVER be coded as both 1 and 1. The positions are mutually exclusive.

> OUTPUT FORMAT:
Return ONLY a valid JSON dictionary in this exact format:
>
>{
  "concept1": 0,
  "concept2": 0
}

>Where:
>- "concept1" is either 0 or 1
>- "concept2" is either 0 or 1

>Do NOT include any other text or explanation.
>Do NOT write "0 or 1" in the output.
>Do NOT wrap the numbers in quotes.
>Do NOT use markdown or backticks.
>Your entire response must be ONLY the JSON object.

## **Reflection on Changing Prompting Strategies For Better Inter-Rater Reliability Scores**

---

In the beginning of the project, our prompting strategy left too much room for interpretation, which made the model inconsistent and lowered our inter-rater reliability. A lot of comments in the dataset were unclear, sarcastic, or not taking a clear stance, and the model kept trying to infer meaning instead of following a strict standard. To improve this, we rewrote the prompt so the model had very clear rules to follow. The most important change was creating a firm guideline for when to assign 0,0 versus 0,1 or 1,0. We specified that a comment must explicitly choose one side of the water debate in order to receive 0,1 or 1,0. If the comment was ambiguous, humorous, meta, or off topic, it had to be coded as 0,0 every time. This single adjustment eliminated most of the inconsistent edge cases that were dragging down our reliability scores. We also refined the definitions of the two main concepts in the debate, clearly explaining what counts as “water is wet” reasoning and what counts as “water is not wet” reasoning. Giving the model sharper conceptual boundaries made its decisions far more consistent.

## **Results:**

---

Our analysis of the 1,000 TikTok comments showed that most users did not agree with the premise that water is wet. Out of the full dataset, about 55 comments clearly supported the idea that water is wet, while about 118 comments argued that water is not wet. This means that among commenters who took a definite position, the argument that water is not wet appeared more than twice as often. These comments generally relied on everyday explanations, such as the idea that wetness describes what happens when water coats another surface. Many commenters repeated versions of the claim that water cannot be wet because it is the substance that makes other things wet. This suggested that viewers tended to draw on intuitive language and familiar logic rather than scientific definitions.

At the same time, a large portion of comments did not engage with the question in any meaningful way. Many users made jokes, tagged friends, reacted to the presenters, or shared unrelated thoughts. There were also many comments comparing the water debate to statements like “fire is hot but not burned,” which echoed the overall structure of the argument without actually taking a stance. Under our coding rules these comments were classified as 0,0, and they confirmed that the TikTok discussion often drifted away from the topic, especially when the prompt encouraged playful or definition based responses.

Looking at engagement patterns added more context to these findings. Although the “water is not wet” stance appeared most often, comments that did not take a clear position were the ones that received the highest average number of likes. Figure 2 shows that neutral or humorous comments tended to draw more positive engagement than comments that argued directly for or against the question. This suggests that TikTok audiences may be more drawn to entertainment and humor than to definitional arguments. Taken together, the results show that while explicit stances leaned toward the idea that water is not wet, the broader discussion was shaped by distraction, playfulness, and a lack of consistent argumentative focus. This helps explain why both human coders and LLMs encountered a large number of comments that fell outside the main categories, and it emphasizes how public discourse on TikTok often blends joking, debate, and drifting attention.
